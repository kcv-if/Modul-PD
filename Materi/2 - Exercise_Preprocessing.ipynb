# Principal Components Analysis

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

data = df_normalized.drop("status_kelulusan", axis=1)
label = df_normalized["status_kelulusan"]

principal_components = pca.fit_transform(data)

principal_components_df = pd.DataFrame(data=principal_components, columns=["PC1", "PC2"])

principal_components_df["label"] = label
principal_components_df.head()
```

```python
# Plot the data in the reduced dimensional space
plt.figure(figsize=(8, 6))
targets = [0, 1]
colors = ['r', 'g']

for target, color in zip(targets, colors):
    indices_to_keep = principal_components_df['label'] == target
    plt.scatter(principal_components_df.loc[indices_to_keep, 'PC1'],
                principal_components_df.loc[indices_to_keep, 'PC2'],
                c=color,
                label=target,
                alpha=0.7)

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Principal Components Analysis')
plt.legend()
plt.show()
```

# Attribute subset selection

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectFromModel
from sklearn.metrics import accuracy_score

X = df_normalized.drop(["id","status_kelulusan"], axis = 1)
y = df_normalized["status_kelulusan"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)

X_train.shape, X_test.shape
```

```python
clf = RandomForestClassifier(random_state=123)
clf.fit(X_train, y_train)

sfm = SelectFromModel(clf, threshold=0.02)
sfm.fit(X_train, y_train)

X_train_selected = sfm.transform(X_train)
X_test_selected = sfm.transform(X_test)
X_train_selected.shape, X_test_selected.shape
```

```python
# Ambil feature importance dari model
importances = clf.feature_importances_

# Buat DataFrame biar rapi
feat_importances = pd.DataFrame({
    "feature": X_train.columns,
    "importance": importances
})

feat_importances
```

```python
# Train and evaluate a classifier with the selected features
clf_selected = RandomForestClassifier(random_state=123)
clf_selected.fit(X_train_selected, y_train)

y_pred = clf.predict(X_test)
y_pred_selected = clf_selected.predict(X_test_selected)

# Evaluate accuracy
accuracy_selected_features = accuracy_score(y_test, y_pred_selected)
accuracy_all_features = accuracy_score(y_test, y_pred)

print(f'Accuracy all features: {accuracy_all_features * 100:.2f}%')
print(f'Accuracy with selected features: {accuracy_selected_features * 100:.2f}%')
```
